```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.path as path
```

```python
vertics = [
    (50, 50),
    (50, 200),
    (150, 500),
    (250, 200),
    (350, 100),
    (150, 150),
    (100, 100),
    (50, 50),
]

codes = [
    path.Path.MOVETO,
    path.Path.LINETO,
    path.Path.CURVE3,
    path.Path.CURVE3,
    path.Path.CURVE4,
    path.Path.CURVE4,
    path.Path.CURVE4,
    path.Path.LINETO,
]

path = path.Path(vertics, codes)
fig, ax = plt.subplots()
patch = patches.PathPatch(path, facecolor="none", lw=2)
ax.add_patch(patch)
ax.set_xlim(0, 400)
ax.set_ylim(0, 600)
plt.gca().set_aspect("equal", adjustable="box")
plt.show()
```
