```python
import matplotlib.pyplot as plt


def midpoint_circle(xc, yc, r):
    points = []

    x = 0
    y = r
    p = 1 - r  # Initial decision parameter

    while x <= y:
        # Add points for all eight octants based on symmetry
        points.append((xc + x, yc + y))
        points.append((xc - x, yc + y))
        points.append((xc + x, yc - y))
        points.append((xc - x, yc - y))
        points.append((xc + y, yc + x))
        points.append((xc - y, yc + x))
        points.append((xc + y, yc - x))
        points.append((xc - y, yc - x))

        x += 1

        if p < 0:
            p += 2 * x + 1
        else:
            y -= 1
            p += 2 * (x - y) + 1

    return points


# Example usage:
xc, yc, r = 5, 5, 5  # Center (5, 5) and radius 5
circle_points = midpoint_circle(xc, yc, r)

# Separate the x and y coordinates for plotting
x_coords = [point[0] for point in circle_points]
y_coords = [point[1] for point in circle_points]

# Plotting the points using matplotlib
plt.figure(figsize=(6, 6))
plt.scatter(x_coords, y_coords, color="blue", s=50)  # Plot points
plt.gca().set_aspect("equal", adjustable="box")  # Equal aspect ratio for circle
plt.title("Midpoint Circle Algorithm")
plt.grid(True)
plt.show()
```
