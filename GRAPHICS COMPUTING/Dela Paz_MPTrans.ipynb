```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

```python
# Prism whose base is rectangle (L = 21 and W = 16) and height is 23 units. (orange)

# Parameters for the rectangular prism
length = 21  # Length of the prism
width = 16  # Width of the prism
height = 23  # Height of the prism

# Define the vertices of the rectangular prism
vertices = np.array(
    [
        [0, 0, 0],  # Vertex 0
        [length, 0, 0],  # Vertex 1
        [length, width, 0],  # Vertex 2
        [0, width, 0],  # Vertex 3
        [0, 0, height],  # Vertex 4
        [length, 0, height],  # Vertex 5
        [length, width, height],  # Vertex 6
        [0, width, height],  # Vertex 7
    ]
)

# Define the 6 faces of the rectangular prism
faces = [
    [vertices[0], vertices[1], vertices[2], vertices[3]],  # Bottom face
    [vertices[4], vertices[5], vertices[6], vertices[7]],  # Top face
    [vertices[0], vertices[1], vertices[5], vertices[4]],  # Front face
    [vertices[1], vertices[2], vertices[6], vertices[5]],  # Right face
    [vertices[2], vertices[3], vertices[7], vertices[6]],  # Back face
    [vertices[3], vertices[0], vertices[4], vertices[7]],  # Left face
]

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# Plot the rectangular prism
ax.add_collection3d(
    Poly3DCollection(
        faces, facecolors="orange", linewidths=1, edgecolors="red", alpha=0.5
    )
)

# Set labels for clarity
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

# Set the limits for the axes
ax.set_xlim([0, length])
ax.set_ylim([0, width])
ax.set_zlim([0, height])

# Display the plot
plt.show()
```

```python
# Translate the prism in the last activity using the parameter [5, 7, 2]

# Define the translation vector
translation_vector = np.array([5, 7, 2])

# Translate the vertices of the rectangular prism
translated_vertices = vertices + translation_vector

# Define the 6 faces of the translated rectangular prism
translated_faces = [
    [translated_vertices[0], translated_vertices[1], translated_vertices[2], translated_vertices[3]],  # Bottom face
    [translated_vertices[4], translated_vertices[5], translated_vertices[6], translated_vertices[7]],  # Top face
    [translated_vertices[0], translated_vertices[1], translated_vertices[5], translated_vertices[4]],  # Front face
    [translated_vertices[1], translated_vertices[2], translated_vertices[6], translated_vertices[5]],  # Right face
    [translated_vertices[2], translated_vertices[3], translated_vertices[7], translated_vertices[6]],  # Back face
    [translated_vertices[3], translated_vertices[0], translated_vertices[4], translated_vertices[7]],  # Left face
]

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# Plot the translated rectangular prism
ax.add_collection3d(
    Poly3DCollection(
        translated_faces, facecolors="orange", linewidths=1, edgecolors="red", alpha=0.5
    )
)

# Set labels for clarity
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

# Set the limits for the axes
ax.set_xlim([0, length + translation_vector[0]])
ax.set_ylim([0, width + translation_vector[1]])
ax.set_zlim([0, height + translation_vector[2]])

# Display the plot
plt.show()
```

```python
# Cube whose initial coordinate located at (4, 8, 9) and the edge is 12. (blue)

# Define the initial corner and edge length
initial_corner = np.array([4, 8, 9])
edge_length = 12

# Calculate the vertices of the cube
vertices = [
    initial_corner,  # (4, 8, 9)
    initial_corner + [edge_length, 0, 0],  # (16, 8, 9)
    initial_corner + [edge_length, edge_length, 0],  # (16, 20, 9),
    initial_corner + [0, edge_length, 0],  # (4, 20, 9),
    initial_corner + [0, 0, edge_length],  # (4, 8, 21),
    initial_corner + [edge_length, 0, edge_length],  # (16, 8, 21),
    initial_corner + [edge_length, edge_length, edge_length],  # (16, 20, 21),
    initial_corner + [0, edge_length, edge_length],  # (4, 20, 21),
]

# Define the faces of the cube using vertex indices
faces = [
    [vertices[0], vertices[1], vertices[2], vertices[3]],  # Bottom
    [vertices[4], vertices[5], vertices[6], vertices[7]],  # Top
    [vertices[0], vertices[1], vertices[5], vertices[4]],  # Front
    [vertices[2], vertices[3], vertices[7], vertices[6]],  # Back
    [vertices[1], vertices[2], vertices[6], vertices[5]],  # Right
    [vertices[4], vertices[7], vertices[3], vertices[0]],  # Left
]

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# Add the faces to the plot
ax.add_collection3d(
    Poly3DCollection(
        faces, facecolors="blue", linewidths=1, edgecolors="red", alpha=0.30
    )
)

# Set labels for clarity
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

# Set limits for the axes to encompass the cube
ax.set_xlim([initial_corner[0], initial_corner[0] + edge_length])
ax.set_ylim([initial_corner[1], initial_corner[1] + edge_length])
ax.set_zlim([initial_corner[2], initial_corner[2] + edge_length])

# Show the plot
plt.show()
```

```python
# Scale the cube in the last activity using the parameter [0, 2, 1]

# Define the scaling vector
scaling_vector = np.array([0, 2, 1])

# Scale the vertices of the cube
scaled_vertices = np.array(vertices) * scaling_vector

# Define the faces of the scaled cube using vertex indices
scaled_faces = [
    [scaled_vertices[0], scaled_vertices[1], scaled_vertices[2], scaled_vertices[3]],  # Bottom
    [scaled_vertices[4], scaled_vertices[5], scaled_vertices[6], scaled_vertices[7]],  # Top
    [scaled_vertices[0], scaled_vertices[1], scaled_vertices[5], scaled_vertices[4]],  # Front
    [scaled_vertices[2], scaled_vertices[3], scaled_vertices[7], scaled_vertices[6]],  # Back
    [scaled_vertices[1], scaled_vertices[2], scaled_vertices[6], scaled_vertices[5]],  # Right
    [scaled_vertices[4], scaled_vertices[7], scaled_vertices[3], scaled_vertices[0]],  # Left
]

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# Add the faces to the plot
ax.add_collection3d(
    Poly3DCollection(
        scaled_faces, facecolors="blue", linewidths=1, edgecolors="red", alpha=0.30
    )
)

# Set labels for clarity
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

# Set limits for the axes to encompass the scaled cube
ax.set_xlim([scaled_vertices[:, 0].min(), scaled_vertices[:, 0].max()])
ax.set_ylim([scaled_vertices[:, 1].min(), scaled_vertices[:, 1].max()])
ax.set_zlim([scaled_vertices[:, 2].min(), scaled_vertices[:, 2].max()])

# Show the plot
plt.show()
```

```python
# Pyramid whose base is rectangle (L = 21 and W = 16) and height is 23 units. (green)

# Parameters for the pyramid
base_width = 21  # Width of the base
base_length = 16  # Length of the base
height = 23  # Height of the pyramid

# Define the vertices of the pyramid
vertices = np.array(
    [
        [-base_length / 2, -base_width / 2, 0],  # Base corner 1
        [base_length / 2, -base_width / 2, 0],  # Base corner 2
        [base_length / 2, base_width / 2, 0],  # Base corner 3
        [-base_length / 2, base_width / 2, 0],  # Base corner 4
        [0, 0, height],  # Apex
    ]
)

# Define the faces of the pyramid (4 triangles + 1 square base)
faces = [
    [vertices[0], vertices[1], vertices[4]],  # Front face
    [vertices[1], vertices[2], vertices[4]],  # Right face
    [vertices[2], vertices[3], vertices[4]],  # Back face
    [vertices[3], vertices[0], vertices[4]],  # Left face
    [vertices[0], vertices[1], vertices[2], vertices[3]],  # Base face
]

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# Plot the pyramid
ax.add_collection3d(
    Poly3DCollection(
        faces, facecolors="green", linewidths=1, edgecolors="red", alpha=0.5
    )
)

# Set labels for clarity
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

# Set the limits for the axes
ax.set_xlim([-base_length, base_length])
ax.set_ylim([-base_width, base_width])
ax.set_zlim([0, height])

# Display the plot
plt.show()
```

```python
#Rotate the pyramid in the last activity using 120 degrees about x, y and z axis.

# Angle of rotation in radians
angle_x = math.radians(120)
angle_y = math.radians(120)
angle_z = math.radians(120)

# Define the rotation matrices
rotation_matrix_x = np.array([
    [1, 0, 0],
    [0, math.cos(angle_x), -math.sin(angle_x)],
    [0, math.sin(angle_x), math.cos(angle_x)]
])

rotation_matrix_y = np.array([
    [math.cos(angle_y), 0, math.sin(angle_y)],
    [0, 1, 0],
    [-math.sin(angle_y), 0, math.cos(angle_y)]
])

rotation_matrix_z = np.array([
    [math.cos(angle_z), -math.sin(angle_z), 0],
    [math.sin(angle_z), math.cos(angle_z), 0],
    [0, 0, 1]
])

# Rotate the vertices of the pyramid
rotated_vertices = np.dot(vertices, rotation_matrix_x)
rotated_vertices = np.dot(rotated_vertices, rotation_matrix_y)
rotated_vertices = np.dot(rotated_vertices, rotation_matrix_z)

# Define the faces of the rotated pyramid
rotated_faces = [
    [rotated_vertices[0], rotated_vertices[1], rotated_vertices[4]],  # Front face
    [rotated_vertices[1], rotated_vertices[2], rotated_vertices[4]],  # Right face
    [rotated_vertices[2], rotated_vertices[3], rotated_vertices[4]],  # Back face
    [rotated_vertices[3], rotated_vertices[0], rotated_vertices[4]],  # Left face
    [rotated_vertices[0], rotated_vertices[1], rotated_vertices[2], rotated_vertices[3]]  # Base face
]

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the rotated pyramid
ax.add_collection3d(Poly3DCollection(rotated_faces, facecolors='green', linewidths=1, edgecolors='red', alpha=.5))

# Set labels for clarity
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set the limits for the axes
ax.set_xlim([-base_length, base_length])
ax.set_ylim([-base_width, base_width])
ax.set_zlim([0, height])

# Display the plot
plt.show()
```
